## Installing and Importing Required Libraries

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.linear_model import LogisticRegression
```

## Loading The Dataset

```python
data = pd.read_csv('Datasets/breast_cancer_bd.csv')
data.head(15)
```

## Data Preprocessing

```python
# Removing Irrelevant Data
data = data.drop('Sample code number', axis=1)
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data['Bare Nuclei'].unique())
```

```python
# Handling Missing Values and Converting Column Data to Numeric
data = data.replace('?', np.nan)

data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'])
data.info()
```

```python
# Replacing Missing Values with Median Values
data['Bare Nuclei'] = data['Bare Nuclei'].fillna(data['Bare Nuclei'].median())
```

## Training The Model

```python
# Splitting Training Data
y = data['Class']
X = data.drop('Class', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
```

```python
# Declaring And Training Logistic Regression Model
lr_model = LogisticRegression()

lr_model.fit(X_train, y_train)
```

## Testing The Model

```python
y_pred_lr = lr_model.predict(X_test)
score = accuracy_score(y_test,y_pred_lr)
print('Accuracy score: {}'.format(score))
```
